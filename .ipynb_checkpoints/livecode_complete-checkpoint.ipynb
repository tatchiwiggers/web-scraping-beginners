## Então vamos começar o nosso scraping!!!

Primeiramente vamos importar as duas bilbiotecas python que vamos utilizar:
requests é a bilbioteca que vamos utilizar para fazer as requisições à pagina do books to scrape e BS a bilbioteca que vamos utilizar para fazer o scraping.
Para executar a celula e fazer o importa dos libs basta usar enter shift ou clicar em executar.

```python
import requests
from bs4 import BeautifulSoup
```

- Aqui nos armazenamos dentro da variavel url a url do site que vamos fazer o scraping.
- Na segunda linha fazemos a requisição à url e guardamos a resposta dentro da variavel response que nos devolve o HTML.
- Nos guardamos dentro da variavael scraped o html parseado pelo BS. 

```python
url = "http://books.toscrape.com/index.html"
response = requests.get(url)
html = response.content
# print(html) # bytes precisamos parsear esses bytes para html
scraped = BeautifulSoup(html, 'html.parser')
```

Agora nos temos um objeto scraped que armazena o HTML parseado da url books to scrape. 

```python
scraped # objeto do <class 'bs4.BeautifulSoup'>
```

percebam que conseguimos buscar elementos da pagina, mas a verdade é que nunca vamos buscar UM ELEMENTO POR VEZ, porque a pagina consiste de elementos repetidos. O HTML em si é todo construido de blocos repetidos... como por exemplo esses cartões que representam cada um, um livro e suas informações. A idia aqui é acessar tudo de uma vez. E como fazemos isso?

### Vamos utilizar um metodo chamado find all e passar as infos do elemento e da classe que queremos acessar. 
- é importante perceber pessoal, o que esse codigo vai retornar. que tipo de objeto ele vai nos retornar:

```python
items = scraped.find_all("article", class_="product_pod")
print(type(items)) # => bs4.element.ResultSet == uma lista de elementos, é uma estrutura de dados criada pelo BS
# que equivale ao array, ou a uma lista em python. percebam: print items
print(items)
```

```python
# para cada item dentro desse array, essa lista de itens
for item in items:
    # eu quero acessar o atributo titulo (que fica dentro de h3 que fica dentro do nosso elemento item)
    # e guardar cada um deles dentro dessa variavel title
    title = item.h3.a["title"]
#     link = item.h3.a["href"]
    # vou immprimir minha variavel title
#     print(title)
#     print(link)

# list comprehension
titles = [ item.h3.a["title"] for item in items ]
# print(titles)
titles = '\n'.join(titles)
print(titles)
```

## queremos os links tb!!
List Comprehension nada mais é do que uma forma rápida e eficiente de criar uma lista a partir de outra lista. Para cada valor de uma lista, você aplica uma operação e coloca todos os resultados em uma nova lista. Isso é List Comprehension.
adicionar no codigo acima:

```python
for item in items:
#     title = item.h3.a["title"]
    link = item.h3.a["href"]
#     print(title)
#     print(link)
    
# list comprehension
links = [ item.h3.a["href"] for item in items ]
# print(links)
links = '\n'.join(links)
print(links)
```

### Vamos agora buscar os preços... temos os titulos, os links...
1. Inspecionar a pagina, e analisar o HTML;
2. buscar o elemento que quero;
3. pensar em como identificar esse elemento de forma única; (class='price_color')

```python
# find all always takes 2 args and the first is always the name of the tag
prices = scraped.find_all('p', class_='price_color')
for price in prices:
    print(price.text)

# list comprehension
prices = [ price.text for price in prices]
prices = '\n'.join(prices)
print(prices)
# print(type(prices))
```

### O problema é que esses números não são inteiros, eles são strings, então eu não consigo manipular esses dados para fazer calculos por exemplo.
- o que preciso fazer aqui é converter esses string para floats. ou seja, convertê-los para números decimais e retirar o sinal de libra.

```python
prices = scraped.find_all('p', class_='price_color')
    
# for price in prices:
#     price = float(price.text.lstrip('£'))
#     print(price)
    
prices = [ float(price.text.lstrip('£')) for price in prices ]
print(prices)
print(prices[0] + prices[1])
```

### Vamos então fazer o scraping de nossos preços associados com cada livro

```python
title_prices = [] # List

# # Select all articles
articles = scraped.find_all("article", class_ = "product_pod")
# print(articles)

for article in articles:
    # para cada article quero:
    title = article.h3.a["title"]
    price = article.find("p", class_="price_color")
    # quero transformar o string em um float
    price_float = float(price.text.lstrip("£"))
    # Crio um Dictionary para cada livro com preço e adiciono à minha lista total prices
    title_prices.append(
        {title: price_float}
    )
#     print(price)
#     print(price_float)
    
print(title_prices)

# optional - if we have time
for title in title_prices:
    for k, v in title.items():
       print(f'{k}: {v}')
```

### uma ultima observação antes de finalizarmos:
o site tem 50 paginas, mas somente fizemos o scraping de uma. A estrutura da url é essa http://books.toscrape.com/catalogue/page-2.html - se quisessemos parsear todos os livros de todas as paginas? como fariamos?

```python
# muito trabalho???
pages = []
for i in range(1, 51):
    pages.append(i)
print(pages)

pages = range(1, 51)
# print(pages)

# vamos usar o f-string - interpolação
for page in pages:
    # print index and element
    # print(f'item {page} está localizado no index {index}')
    link = f'http://books.toscrape.com/catalogue/page-{page}.html'
    print(link)
    
# ou melhor -> vamos usar um list comprehension

list = [f'http://books.toscrape.com/catalogue/page-{page}.html' for page in pages]
# print(list)
list = '\n'.join(list)
print(list)
```
